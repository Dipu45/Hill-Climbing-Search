```python
import random
import copy
import math

# generate random number between 0 to 1
random.seed(100)
n = random.random()
print(n)

# this cell will print different random numbers on different runs
for i in range(5):
    print(random.random())

# this cell will print same random numbers on different runs
random.seed(100)
for i in range(5):
    print(random.random())

# accept a bad move with probability P
random.seed(100) 
P = 0.56
X = random.random()
print(X)
if X <= P:
    print('accept the bad move')
else:
    print('do not accept')

print(math.exp(5))

# list copy
curr = [[3, 1, 2], 
        [0, 4, 5], 
        [6, 7, 8]]

#make a deep copy of the 2D list
neighbour = copy.deepcopy(curr)
#modify some values in the deep copied list
neighbour[1][0] = 3
neighbour[0][0] = 0

#print the original list
for i in range(3):
    print(curr[i])

#print the modified list
for i in range(3):
    print(neighbour[i])

#define a schedule function that returns a value based on the input parameter
def schedule(t):
    if t >= 500:
        return 0
    else:
        return 500 - t


#test the schedule function with different input parameters
print(schedule(1))
print(schedule(50))


#define start and goal states for a puzzle problem
start_state = [[3, 1, 2], 
               [0, 4, 5], 
               [6, 7, 8]]
goal_state = [[0, 1, 2], 
              [3, 4, 5], 
              [6, 7, 8]]

#modify a value in the start state
start_state[0][2] = 3
print(start_state[0][2])

#make a deep copy of the start state
state1 = copy.deepcopy(start_state)

def h1(state):
    # compute number of misplaced tiles heuristic
    num_misplaced_tiles = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != goal_state[i][j]:## if the value of the current cell in the state does not match the value of the corresponding cell in the goal state
                num_misplaced_tiles += 1 ## increment the value of num_misplaced_tiles by 1
    return num_misplaced_tiles ## return the final value of num_misplaced_tiles as the heuristic value

def h2(state):
    # compute manhattan distance heuristic
    manhattan_distance = 0
    for i in range(3):
        for j in range(3):
          # check if the value in the current cell is not 0
            if state[i][j] != 0:
              # compute the correct row and column of the current value
                row = (state[i][j] - 1) // 3
                col = (state[i][j] - 1) % 3
                # calculate the manhattan distance between the current cell and its correct position
                manhattan_distance += abs(row - i) + abs(col - j)
    return manhattan_distance ## return the total manhattan distance

def get_lowest_cost_successor(state):
    # get all neighbors of the current state
    neighbors = get_neighbors(state)

    # calculate the cost of each neighbor
    costs = [h2(neighbor) for neighbor in neighbors]

    # find the index of the neighbor with the lowest cost
    lowest_cost_index = costs.index(min(costs))

    # get the neighbor with the lowest cost
    neighbor = neighbors[lowest_cost_index]

    # return the neighbor and its heuristic value
    return neighbor, h2(neighbor)


def get_neighbors(state):
    # find the location of the blank tile
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                blank_row, blank_col = i, j

    # generate all possible neighbors
    neighbors = []
    if blank_row > 0:
        # move the blank tile up
        neighbor = copy.deepcopy(state)
        neighbor[blank_row][blank_col] = neighbor[blank_row - 1][blank_col]
        neighbor[blank_row - 1][blank_col] = 0
        neighbors.append(neighbor)
    if blank_row < 2:
        # move the blank tile down
        neighbor = copy.deepcopy(state)
        neighbor[blank_row][blank_col] = neighbor[blank_row + 1][blank_col]
        neighbor[blank_row + 1][blank_col] = 0
        neighbors.append(neighbor)
    if blank_col > 0:
        # move the blank tile left
        neighbor = copy.deepcopy(state)
        neighbor[blank_row][blank_col] = neighbor[blank_row][blank_col - 1]
        neighbor[blank_row][blank_col - 1] = 0
        neighbors.append(neighbor)
    if blank_col < 2:
        # move the blank tile right
        neighbor = copy.deepcopy(state)
        neighbor[blank_row][blank_col] = neighbor[blank_row][blank_col + 1]
        neighbor[blank_row][blank_col + 1] = 0
        neighbors.append(neighbor)

    return neighbors

def cost(state):
    return h1(state) + h2(state)


def hill_climbing(start_state):
    curr = start_state
    itr = 0
    while True:
        itr += 1
        if itr > 300: 
            break
        neighbor, h = get_lowest_cost_successor(curr)
        print(neighbor, h)
        if h >= h2(curr):
            return curr
        curr = neighbor
    print(curr)



hill_climbing(start_state)
```
